```python
import sympy
```

# Create function, var, matrix, ...

```python
x, y = sympy.symbols('x y') #creat var
f = x**2 + 3*x + 3 * y**2 #creat f

sympy.pprint(f)
print("calculate f with x = 3: ",f.subs(x, 3))
print("calculate f with x = 3, y = 4: ",f.subs([(x, 3), (y, 4)]))
```

```python
matrix = sympy.Matrix([[2, -3], [1, 5]])
matrix_var = sympy.Matrix([[2, -x], [1, 5]])

sympy.pprint(matrix)
print("---------------")
sympy.pprint(matrix_var)
```

# Convert

```python
str_expr = "x**2 + 3*x - 1/2" # convert from string to sympy
f = sympy.sympify(str_expr)
sympy.pprint(f)
print('f with x = 3: ', f.subs(x, 3))
```

# Derivative 

```python
x = sympy.symbols('x')
f = x**2 + 3*x

derivative = sympy.diff(f, x, 1)
second_derivative = sympy.diff(f, x, 2)

print("Derivative: ", end="")
sympy.pprint(derivative)
print("Second derivative: ", end="")
sympy.pprint(second_derivative)
print("Derivative f(x) with x = 3: ", derivative.subs(x, 3))
```

## Derivative Matrix

```python
matrix = sympy.Matrix([[2, -x], [1, 3]])
sympy.diff(matrix, x)
```

# Integration

```python
integral = sympy.integrate(sympy.cos(x), x)
print('indefinite integral of cos(x): ', integral)
```

```python
integral = sympy.integrate(sympy.exp(-x), (x, 0, sympy.oo))
print('indefinite integral of exp(x): ', integral)
```

# Limit

```python
limit = sympy.limit(1/x, x, sympy.oo) #litmit of 1/x to oo
print("Litmit of 1/x to oo is: ", limit)
```

```python
limit3 = sympy.limit(sympy.cos(x)/x, x, 0) #litmit of cos(x) to oo
print("Litmit of sin(x) to 0 is: ", limit3)
```

# Series Expansion

```python
series = sympy.series(sympy.sin(x), x) # compute Taylor series of sin(x)
sympy.pprint(series)
```

```python
series = sympy.series(sympy.sin(x), x, 0, 8) # compute Taylor series of sin(x) end at n from 0 to 8
sympy.pprint(series)
```

# Jacobian matrix

```python
X = sympy.Matrix([x**2 + y, x**3 - y, x + y, 2*x**2 +3*y**3])
Y = sympy.Matrix([x, y])
X.jacobian(Y)
```
